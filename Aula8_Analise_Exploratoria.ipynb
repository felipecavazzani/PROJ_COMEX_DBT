```python
#Criando nosso DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a Receita total?
df["Valor Venda"].sum()
```

```python
#Custo Total
df["custo"] = df["Custo Unit√°rio"].mul(df["Quantidade"]) #Criando a coluna custo
```

```python
df.head(1)
```

```python
#Qual o custo Total?
round(df["custo"].sum(), 2) #round arredonda para 2 casas decimais
```

```python
#Agora que temos a receita e custo total, podemos achar o lucro total
#Vamos criar uma coluna de Lucro que ser√° : Receita - Custo
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
round(df["lucro"].sum(),2)
```

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

**Agora, queremos saber a m√©dia do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em num√©rica**

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#M√©dia do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

**Missing Values**

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

**Se quiser saber o Lucro por Ano e Por Marca?**

```python
#Vamos Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#Formatando a exibi√ß√£o dos valores
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o INDEX
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gr√°fico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Gr√°fico do Lucro por ano
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Gr√°fico em linhas do lucro x mes em 2009
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x M√™s")
plt.xlabel("M√™s")
plt.ylabel("Lucro");
```

```python
#Gr√°fico em barras do Lucro x Marca
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal'); #eixo x na horizontal
```

```python
#Gr√°fico em barras do Lucro x Classe
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
